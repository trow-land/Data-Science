```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# defining a dataset

np.random.seed(0)

# generate years of experience values between 0 and 10
x = np.random.uniform(0, 10, 100).reshape(-1, 1)

# Generate target salary with some noise
true_slope = 3.5
true_intercept = 20
noise = np.random.normal(0, 3, x.shape[0]).reshape(-1, 1)

y = true_slope * x + true_intercept + noise
```

```python
# to turn the examples into a design matrix add a column of ones
X = np.column_stack((np.ones(x.shape[0]), x))
```

```python
# turn y into a column vector 
y = y.reshape(-1, 1)
y[:5]
```

Normal Equation$$
\theta = (X^T X)^{-1} X^T y
$$

```python
# Normal equation
X_T = X.transpose()             # (2,4)
XTX = np.matmul(X_T, X)         # (2,2)
XTX_inv = np.linalg.inv(XTX)    # (2,2)
XTy = np.matmul(X_T, y)         # (2,1)
theta = np.matmul(XTX_inv, XTy) # (2,1)
```

```python
print("Parameters θ:", theta)
```

```python
# making predictions 
y_pred = theta[0][0] + theta[1][0] * x
```

```python
# plotting the results
plt.scatter(x, y, color= 'blue', label='Original Data')
plt.plot(x, y_pred, color='red', label='Fitted Line')
plt.xlabel('x')
plt.ylabel('y', rotation=0)
plt.legend()
plt.show()
```
