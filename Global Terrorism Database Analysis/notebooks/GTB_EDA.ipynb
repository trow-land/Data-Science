```python
columns_to_keep = [
    'iyear',
    'imonth',
    'iday',
    'approxdate',
    'region_txt',
    'country_txt',
    'provstate',
    'city',
    'latitude',
    'longitude',
    'gname',
    'attacktype1_txt',
    'success',
    'suicide',
    'nkill',
    'nwound'
]


df = df[columns_to_keep]
```

```python
df.rename(columns={'iyear': 'year', 'imonth': 'month', 'iday': 'day'})
```

```python
df = df.rename(columns={'iyear': 'year', 'imonth': 'month', 'iday': 'day'})

# drop rows with a nonsensical date
df = df[(df['day'] != 0) & (df['month'] != 0)]

# combine into a single date columns
df['date'] = df['day'].astype(str) + '/' + df['month'].astype(str) + '/' + df['year'].astype(str)
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')
```

```python
df['suicide'].value_counts()
```

```python
suicide_df = df[['year', 'suicide']]
```

```python
grouped_sdf = suicide_df.groupby('year').sum().reset_index()
```

```python
grouped_sdf
```

```python
sns.lineplot(data=grouped_sdf, x='year', y='suicide')
```
