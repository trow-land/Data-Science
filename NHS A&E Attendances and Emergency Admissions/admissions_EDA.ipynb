## Insights

### Temporal trends

```python
df.columns.values
```

```python
monthly_cols = ['Period', 'A&E attendances Type 1', 'A&E attendances Type 2',
       'A&E attendances Other A&E Department',
       'A&E attendances Booked Appointments Type 1',
       'A&E attendances Booked Appointments Type 2',
       'A&E attendances Booked Appointments Other Department']

monthly_data = df[monthly_cols]
```

```python
total_monthly_data = monthly_data.groupby(['Period'], as_index=False).sum()
total_monthly_data['Period_label'] = total_monthly_data['Period'].dt.strftime('%B %Y')
total_monthly_data
```

```python
total_monthly_data.columns.values
```

```python
total_monthly_data['Period'].unique()
```

```python
month_order = ['APRIL-2024', 'MAY-2024', 'JUNE-2024', 'JULY-2024', 'AUGUST-2024', 'SEPTEMBER-2024', 'OCTOBER-2024', 'NOVEMBER-2024', 'DECEMBER-2024', 'JANUARY-2025', 'FEBRUARY-2025', 'MARCH-2025']
total_monthly_data.sort_values(by='Period')
```

```python
plt.figure().set_figwidth(20)
plt.plot(total_monthly_data['Period_label'], total_monthly_data['A&E attendances Type 1'])
plt.ylim(1000000, 1600000)
plt.ylabel("A&E Attendances (Million)")
plt.title("Monthly A&E Type 1 Attendances between April 2024 and March 2025")
plt.show()
```

```python
# attendances over 4 hours

four_hours_cols = ['Attendances over 4hrs Type 1', 'Attendances over 4hrs Type 2',
       'Attendances over 4hrs Other Department',
       'Attendances over 4hrs Booked Appointments Type 1',
       'Attendances over 4hrs Booked Appointments Type 2',
       'Attendances over 4hrs Booked Appointments Other Department', 'Period']
four_hours = df[four_hours_cols].groupby(['Period'], as_index=False).sum()
```

```python
four_hours['Period_label'] = four_hours['Period'].dt.strftime("%B %Y")
four_hours
```

```python
plt.figure().set_figwidth(20)
plt.plot(four_hours['Period_label'],four_hours['Attendances over 4hrs Type 1'])
plt.ylim(400000, 750000)
plt.title("Total attendances per month at a Type 1 A&E Facility that were more than 4 hours")
plt.ylabel("Attendances over 4hrs")
```

```python
# combined plots

combined = df.copy()
combined = combined.groupby(['Period'], as_index=False).sum()
combined['Period_label'] = combined['Period'].dt.strftime("%B %Y")
combined
```

```python

x = combined['Period_label']

y_cols_1 = ['A&E attendances Type 1',
       'A&E attendances Other A&E Department']

y_cols_2 = ['A&E attendances Type 2',
       'A&E attendances Booked Appointments Type 1']

fig, ax = plt.subplots(2, 1, figsize=(20, 10))
for col in y_cols_1:
    ax[0].plot(x, combined[col], label=col)
    ax[0].set_title("Type 1 Attendances vs Other A&E Dept Attendances")
    ax[0].set_ylabel("Attendances (Million)")
    ax[0].legend()
for col in y_cols_2:
    ax[1].plot(x, combined[col], label=col)
    ax[1].set_ylabel("Attendances")
    ax[1].set_title("Type 2 Attendances vs Booked Type 1")
    ax[1].legend()

plt.show()
```
