```python
# The numerical columns need to have the special characters removed to avoid NAN when changing datatype

num_column_list = ['discounted_price', 'actual_price', 'discount_percentage','rating_count', 'rating']

numerical_columns = df[num_column_list]
numerical_columns
```

```python
numerical_columns = numerical_columns.apply(lambda x: x.astype(str)
                                            .str.replace(r'[‚Çπ,%]', '', regex=True) 
                                            .str.replace(',', '')) 
numerical_columns
```

```python
# Converting numerical columns to correct dtype for now

numerical_columns['discounted_price'] = pd.to_numeric(numerical_columns['discounted_price'], errors='coerce')
numerical_columns['actual_price'] = pd.to_numeric(numerical_columns['actual_price'], errors='coerce')
numerical_columns['discount_percentage'] = pd.to_numeric(numerical_columns['discount_percentage'], errors='coerce')
numerical_columns['rating_count'] = pd.to_numeric(numerical_columns['rating_count'], errors='coerce')
numerical_columns['rating'] = pd.to_numeric(numerical_columns['rating'], errors='coerce')
```

```python
# check for nan values post conversion
numerical_columns.isna().sum()
```

```python
# The nan values were from the original data, they will need to be removed once the numerical and text columns have been merged
```

```python
# Check lengths match
print(len(df), len(numerical_columns))
```

```python
# Converty from indian rupees to gbp
# 1 INR = 0.0095 GBP

numerical_columns['discounted_price'] = (numerical_columns['discounted_price'] * 0.0095).round(2)
numerical_columns['actual_price'] = (numerical_columns['actual_price'] * 0.0095).round(2)
numerical_columns
```

```python
# create df of the text columns
text_columns = df.loc[:, ~df.columns.isin(num_column_list)]
text_columns.head()
```

```python
# Merge the text columns and the numerical columns on the original index

merged = text_columns.merge(numerical_columns, left_index=True, right_index=True)
```

```python
merged.head()
```

```python
# Check for nan values
merged.isna().sum()
```

```python
cleaned = merged.dropna()
print(len(merged), len(cleaned))
```
